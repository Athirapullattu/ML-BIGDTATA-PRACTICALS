```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
rng=np.random.RandomState(1)
X=np.dot(rng.rand(2,2),rng.randn(2,200)).T
X
```

```python
X_df=pd.DataFrame(X,columns=['col1','col2'])
X_df
```

```python
sns.lmplot(x='col1',y='col2',data=X_df)
```

```python
from sklearn.decomposition import PCA
model=PCA(n_components=1)
model.fit(X_df)
```

```python
X_1D=model.transform(X_df)
X_1D
```
